```python
import ISLP
colleges = ISLP.load_data("College")
```

```python
# For ease of use, having "." in a columnname is not nice. Rename all columns with a "." in their columnname by exchanging it with a "_". (Hint: DataFrame.rename)
import pandas as pd


colleges_pd = pd.DataFrame(colleges)

for column in colleges_pd.columns:
    new_col = column.replace('.', '_')
    # https://stackoverflow.com/a/66607872
    colleges_pd.columns = colleges_pd.columns.str.replace(column, new_col, regex=False)

print(colleges_pd)
```

```python
# Create a new variable "Acceptance_Rate" that calculates the number of students accepted in % based on the number of applications and add it to the dataframe.

colleges_pd["Acceptance_Rate"] = (colleges_pd["Accept"]/colleges_pd["Apps"]*100)
print(colleges_pd)
```

```python
# Sort the Dataframe by the variables "Private" and "Apps", consecutively in an ascending manner.

colleges_pd = colleges_pd.sort_values(by = ['Private', 'Apps'], ascending=[True, True])
print(colleges_pd)
```

```python
# Create a new binary qualitative variable, called "Elite", by binning the "Top10perc" variable into two groups based on whether or not the proportion of students coming from
# the top 10% of their high school classes exceeds 50%. (Hint: pd.cut function)

# https://stackoverflow.com/a/54343112

bins = [0, 51, 100]
labels = ["No", "Yes"]
colleges_pd['Elite'] = pd.cut(colleges_pd['Top10perc'], bins=bins, labels=labels)
colleges_pd.tail()
```

```python
# Use the describe() method of to produce a numerical summary of the variables in the data set.

colleges_pd.describe()
```

```python
# Group the data by the variable "Private" and investigate the mean for the variables "Acceptance_Rate", "Room_Board", "Books", "Phd" and "Graduation_Rate". What can you see?

ar = colleges_pd.groupby(['Private'])["Acceptance_Rate"].mean()
rb = colleges_pd.groupby(['Private'])["Room_Board"].mean()
b = colleges_pd.groupby(['Private'])["Books"].mean()
phd = colleges_pd.groupby(['Private'])["PhD"].mean()
gr = colleges_pd.groupby(['Private'])["Grad_Rate"].mean()
print(ar, rb, b, phd, gr)
```

```python
# Group the data by "Private" AND "Elite" and consider the same variables. Can you see a difference?

ar = colleges_pd.groupby(['Private', 'Elite'])["Acceptance_Rate"].mean()
rb = colleges_pd.groupby(['Private', 'Elite'])["Room_Board"].mean()
b = colleges_pd.groupby(['Private', 'Elite'])["Books"].mean()
phd = colleges_pd.groupby(['Private', 'Elite'])["PhD"].mean()
gr = colleges_pd.groupby(['Private', 'Elite'])["Grad_Rate"].mean()
print(ar, rb, b, phd, gr)
```
